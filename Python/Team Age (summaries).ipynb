```python
# Conserver uniquement les 3 colonnes spécifiées
df5 = df5[['season', 'team', 'age']]
df5.head()
print(df5)
```

```python
# Changer le datatype
df5['season'] = pd.to_datetime(df5['season'], format='%Y')

# Supprimer les années NaN
filtered_df5 = df5.dropna(subset=['season'])

print(filtered_df5)
```

```python
# Supprimer les valeurs NaN
df5.dropna(inplace=True)
print(df5)
```

```python
df5.to_csv('Team Age (summaries).csv', index=False)
```

```python
league_average_data = df5.loc[df5['team'] == 'League Average']
print(league_average_data)
```

```python
# Calculer la moyenne d'âge par saison
average_age_per_season = df5.groupby('season')['age'].mean().reset_index().round(2)

# Renommer les colonnes pour plus de clarté
average_age_per_season.columns = ['season', 'average_age']

# Configurer pandas pour afficher toutes les lignes
pd.set_option('display.max_rows', None)

print(average_age_per_season)
```

```python
duplicates = league_average_data[league_average_data.duplicated(['season'], keep=False)]
print("Doublons trouvés :\n", duplicates)
```

```python
# Filtrer les lignes excluant "League Average"
filtered_data = df5.loc[df5['team'] != 'League Average']

# Calculer la moyenne d'âge par saison
league_average_data = filtered_data.groupby('season')['age'].mean().reset_index()

# Renommer les colonnes pour plus de clarté
league_average_data.columns = ['season', 'average_age']

# Supprimer les doublons s'il y en a
league_average_data = league_average_data.drop_duplicates(subset=['season'])

# Créer le graphique
plt.figure(figsize=(10, 6))  
plt.plot(league_average_data['season'], league_average_data['average_age'], marker='o', linestyle='-')  
plt.title("Âge moyen des joueurs de la ligue par année")
plt.xlabel('Années')
plt.ylabel('Âge moyen')
plt.xticks(rotation=45)  
plt.grid(True)  
plt.tight_layout()  
plt.show()
```
